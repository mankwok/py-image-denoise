# Adding Noise to Image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from skimage import data, img_as_float
from skimage.measure import compare_ssim as ssim
from skimage.util import random_noise
from skimage.color import rgb2gray
```

## ISO Noise

```python
img_100_rgb = cv2.imread('D:/Chrome_downnloads/IMG_1927a.jpg')[150:300, 50:300]
img_3200_rgb = cv2.imread('D:/Chrome_downnloads/IMG_1928a.jpg')[150:300, 50:300]
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

a = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(cv2.cvtColor(img_100_rgb, cv2.COLOR_BGR2RGB))
a.set_title('Image at ISO 100')

a = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(cv2.cvtColor(img_3200_rgb, cv2.COLOR_BGR2RGB))
a.set_title('Image at ISO 3200')
```

## Salt-and-pepper Noise

```python
original = img_as_float(data.coffee()[100:250, 50:300])
noisy = random_noise(original, mode='s&p')
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

a = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(original)
a.set_title('Original Image')

a = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(noisy)
a.set_title('Salt-and-pepper Noise')
```

## Shot Noise

```python
original = img_as_float(data.clock()[100:300, 50:300])
noisy = random_noise(original, mode='Poisson')
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

a = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(original)
a.set_title('Original Image')

a = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(noisy)
a.set_title('Shot(Poisson) Noise')
```
